```python
import numpy as np
import pandas as pd

def readCSV(filename):
    df=pd.read_csv(filename)
    return df

df=readCSV('DataFiles/Income.csv')

y=df.values[4]
x=df.columns
```

```python
years=[]
for i in range(4,len(x)):
    years.append(x[i])
years
```

```python
income=[]
for i in range(4,len(y)):
    income.append(y[i])
income
```

```python
plt.xlabel('Year')
plt.ylabel('Income')
plt.title('income of california')  # adding title
plt.grid(True)
plt.plot(years,income)
plt.savefig('Figures/bargraphOfCalifornia.png')
plt.show()
```

```python
width=1/2
plt.bar(years,income,width,color='blue')
plt.savefig('Figures/bargraphcolorblue.png')
```

```python
width=1/2
plt.title('income of california')
plt.bar(years,income,width,color='green')
plt.savefig('Figures/bargraphcolorgreen.png')
```

### Scatter Plot
### Problem 4:
#### Problem Statement:
 generate a scatter plot for 50 random numbers
#### Constraint:
#### Test Case 1:
#### Test Case 2:
#### Test Case 3:

```python
n=50
x=np.random.rand(n)                                 #generate 50 random numbers
y=np.random.rand(n)                                 #generate 50 random numbers
colorlist = np.random.rand(n)                       #generate 50 random colors
arealist = np.pi * (15 * np.random.rand(50))**2     #generate 50 random areas of circles   ,np.random.rand(50) or we can fix=50
plt.scatter(x,y,s=arealist,c=colorlist,alpha=1)                             # if we fix '50' all circles will be in same size
plt.show()
```
